# Problem:
Based on the table shown below, find out the price of a home that has:

<b>3000 sqr ft area, 3 bedrooms, and 40 year old.</b>\
<b>2500 sqr ft area, 4 bedrooms, and 5 year old.</b>

## Import Libraries

````python
import pandas as pd
import numpy as np
from sklearn import linear_model
````

## Load the Data

````python
df=pd.read_csv('D:/Data_Science/My Github/3. Machine-Learning-with-Python/2. Linear_Regression_with_Multiple_Variables/Example/homeprices.csv')
df
````

## Data Preprocessing: Handling  Missing Data

````python
# calculate the median for NaN Data
import math
median_bedrooms = math.floor(df.bedrooms.median())
median_bedrooms
````

````python
# import median_bedrooms into the data by using fillna
df.bedrooms = df.bedrooms.fillna(median_bedrooms)
df
````

## Linear Regression Using Multiple Variables

````python
reg = linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df.price)
````

## Calculate the Coefficients

````python
print('m1 = %s m2 = %s m3 = %s b = %s' %(reg.coef_[0],reg.coef_[1],reg.coef_[2],reg.intercept_))
````

**y = m1 $*$ <font color=red>x1</font> + m2 $*$ <font color=red>x2</font> + m3 $*$ <font color=red>x3</font> + b**

## Prediction
**Predict the price of home by Linear Regression model with 3000 sqr ft area, 3 bedrooms,  40 year old**

````python
reg.predict([[3000,3,40]])
````
